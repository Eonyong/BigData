```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from xgboost import XGBClassifier, plot_importance, plot_tree
```

```python
train = pd.read_csv('/settle/train.csv', index_col = 0)
test = pd.read_csv('/settle/test.csv', index_col = 0)
sample = pd.read_csv('/settle/sample_submission.csv', index_col = 0)
```

```python
column_number = {}
for i, column in enumerate(sample.columns):
    column_number[column] = i
    
def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

```python
train_x = train.drop(columns=['type', 'type_num'], axis=1)
train_y = train['type_num']
test_x = test
```

```python
%%time
xgb_clf = XGBClassifier(n_estimators=1300,
                        max_depth = 7,
                        learning_rate = 0.0001,
                        subsample = 7,
                        colsample_bytree = 0.78,
                        colsample_bylevel = 0.8,
                        min_child_weight = 0.5,
                        gamma = 0.25,
                        reg_lambda = 5.0,
                        n_jobs=-1, random_state=0, tree_method='gpu_hist')
xgb_clf.fit(train_x, train_y)

plot_important(xgb_clf)
plt.show()
```

```python
plot_tree(xgb_clf)
plt.show()
```

```python
xgb_pred = xgb_clf.predict_proba(test_x)
```

```python
submission = pd.DataFrame(data=xgb_pred, columns=sample.columns, index=sample.index)
submission.to_csv('/settle/submission3.csv', index=True)
```
